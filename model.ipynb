```python
numeric_col = df.select_dtypes(exclude=['str','bool']).columns

for col in numeric_col:
    plt.figure(figsize=(12,6))
    plt.title(f'Distribuition of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    sns.histplot(df[col],kde=True,bins=30,color='blue')
    plt.show()
```

```python
for col in numeric_col:
    df_sorted = df.sort_values(by=col,ascending=False).head(10).sort_values(by=col, ascending=False)
    plt.figure(figsize=(12,6))
    plt.title(f'Ranking 10 best {col}')
    plt.xlabel(col)
    plt.ylabel('Channel')
    sns.barplot(data=df_sorted, x=col,y='Channel')
    plt.show()
```

```python
for col in numeric_col:
    plt.title(f'Distribuition of parterned based on {col}')
    sns.barplot(data=df,x='Partnered',y=col,hue='Mature')
    plt.show()
```

```python
for col in numeric_col:
    plt.title(f'BoxPlot of {col}')
    sns.boxplot(data=df,y=col,hue='Partnered')
    plt.show()
```

```python
plt.title('Distribuition of language')
plt.figure(figsize=(19,6))
plt.xlabel('Language')
plt.ylabel('Count')
sns.countplot(data=df,x='Language')
plt.show()
```

```python
plt.title('Distribuition of Mature')
plt.xlabel('Mature')
plt.ylabel('Count')
sns.countplot(data=df,x='Mature')
plt.show()
```
